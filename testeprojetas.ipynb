```python
### DESENVOLVIMENTO VIEW 1 ###


#CRIANDO COLUNA ROTA NO DF VRA
dfVRA_temp0 = spark.sql("""
        SELECT 
            Chegada_Prevista, Chegada_Real, Codigo_Autorizacao, Codigo_Justificativa, Codigo_Tipo_Linha, Nome_Aerodromo_Destino, Estado_Aerodromo_Destino, UF_Aerodromo_Destino, ICAO_Aerodromo_Destino, 
            Nome_Aerodromo_Origem, Estado_Aerodromo_Origem, UF_Aerodromo_Origem, ICAO_Aerodromo_Origem, CONCAT(ICAO_Aerodromo_Origem, ICAO_Aerodromo_Destino) AS Rota, ICAO_Empresa_Aerea, Numero_Voo,
            Partida_Prevista, Partida_Real, Situacao_Voo       
        FROM 
            vra_final
""")

#TRANSFORMANDO dfVRA_temp0 TEM TEMPVIEW
dfVRA_temp0.createOrReplaceTempView("dfVRA_temp0")





#CONTANDO QUANTAS VEZES CADA ROTA FOI FEITA POR CADA EMPRESA AEREA
dfVRA_temp1 = spark.sql("""
    SELECT
        ICAO_Empresa_Aerea, Nome_Aerodromo_Destino, Estado_Aerodromo_Destino, UF_Aerodromo_Destino, ICAO_Aerodromo_Destino, Nome_Aerodromo_Origem, Estado_Aerodromo_Origem, UF_Aerodromo_Origem, 
        ICAO_Aerodromo_Origem, Rota, count(Rota) AS quantidade
    FROM
        dfVRA_temp0
    GROUP BY
        ICAO_Empresa_Aerea, Nome_Aerodromo_Destino, Estado_Aerodromo_Destino, UF_Aerodromo_Destino, ICAO_Aerodromo_Destino, Nome_Aerodromo_Origem, Estado_Aerodromo_Origem, UF_Aerodromo_Origem, 
        ICAO_Aerodromo_Origem, Rota
""")

#TRANSFORMANDO dfVRA_temp1 TEM TEMPVIEW
dfVRA_temp1.createOrReplaceTempView("dfVRA_temp1")





#DESCOBRINDO QUAL O VALOR MAXIMO DA QUANTIDADE DE VEZES QUE A ROTA MAIS FREQUENTE FOI FEITA
dfVRA_temp2 = spark.sql("""
    SELECT
        ICAO_Empresa_Aerea, MAX(quantidade) as max_quantidade
    FROM
        dfVRA_temp1
    GROUP BY
        ICAO_Empresa_Aerea

""")

dfVRA_temp2.createOrReplaceTempView("dfVRA_temp2")





#CRIANDO UMA TABELA RESULTADO COM A EMPRESA AEREA, SUA RESPECTIVA ROTA MAIS FREQUENTE E A QUANTIDADE DE VEZES QUE ELA FOI FEITA
dfVRA_temp3 = spark.sql("""
    SELECT
        dfVRA_temp1.ICAO_Empresa_Aerea, dfVRA_temp1.Rota, dfVRA_temp1.quantidade, dfVRA_temp1.Nome_Aerodromo_Origem, dfVRA_temp1.Estado_Aerodromo_Origem, dfVRA_temp1.UF_Aerodromo_Origem,
        dfVRA_temp1.ICAO_Aerodromo_Origem, dfVRA_temp1.Nome_Aerodromo_Destino, dfVRA_temp1.Estado_Aerodromo_Destino, dfVRA_temp1.UF_Aerodromo_Destino, dfVRA_temp1.ICAO_Aerodromo_Destino     
    FROM
        dfVRA_temp1
    RIGHT JOIN
        dfVRA_temp2 ON dfVRA_temp1.ICAO_Empresa_Aerea = dfVRA_temp2.ICAO_Empresa_Aerea
    WHERE
        dfVRA_temp1.quantidade = dfVRA_temp2.max_quantidade
    ORDER BY 
        ICAO_Empresa_Aerea

""")

dfVRA_temp3.createOrReplaceTempView("dfVRA_temp3")





#SELECIONANDO COLUNAS DESEJADAS PARA A VIEW E TRAZENDO SOMENTE AS INFORMACOES DOS DADOS CONTIDOS NA TABELA AIRCIA
df_view1 = spark.sql("""
    SELECT
        aircia.Razao_Social,dfVRA_temp3.quantidade AS Qtd_Rota_Realizada, dfVRA_temp3.Nome_Aerodromo_Origem, dfVRA_temp3.ICAO_Aerodromo_Origem, dfVRA_temp3.Estado_Aerodromo_Origem, dfVRA_temp3.UF_Aerodromo_Origem, 
        dfVRA_temp3.Nome_Aerodromo_Destino, dfVRA_temp3.ICAO_Aerodromo_Destino, dfVRA_temp3.Estado_Aerodromo_Destino, dfVRA_temp3.UF_Aerodromo_Destino
    FROM
        dfVRA_temp3
    RIGHT JOIN
        aircia ON dfVRA_temp3.ICAO_Empresa_Aerea = aircia.ICAO


""")


#VIEW 1

#EMPRESAS AEREAS POSSUEM DADOS NULOS DEVIDO AO FATO DE NAO TEREM DADOS DE VIAGENS NA TABELA DE VRA
#EMPRESAS AEREAS DUPLICADAS SAO DEVIDO AO FATO DE TEREM MAIS DE UMA ROTA REALIZADAS COM A MESMA QUANTIDADE MAXIMA
df_view1.show()



```

```python
### DESENVOLVIMENTO VIEW 2 ###


#USANDO A TEMPVIEW vra_final CRIADA ANTERIORMENTE, COM OS DADOS DE VRA E DA API UNIFICADOS

#CRIANDO TABELA COM A QUANTIDADE DE PARTIDAS DE CADA AEROPORTO
spark.sql("""
    SELECT 
        Nome_Aerodromo_Origem, ICAO_Aerodromo_Origem, ICAO_Empresa_Aerea, COUNT(ICAO_Aerodromo_Origem) AS quantidade_partidas_do_aeroporto
    FROM 
        vra_final
    GROUP BY
        Nome_Aerodromo_Origem, ICAO_Aerodromo_Origem, ICAO_Empresa_Aerea
""").createOrReplaceTempView("tb_quantidade_partidas")


#CRIANDO TABELA COM A QUANTIDADE EM QUE CADA AEROPORTO TEVE COMO DESTINO
spark.sql("""
    SELECT 
        Nome_Aerodromo_Destino, ICAO_Aerodromo_Destino, ICAO_Empresa_Aerea, COUNT(ICAO_Aerodromo_Destino) AS quantidade_destino_do_aeroporto
    FROM 
        vra_final
    GROUP BY
        Nome_Aerodromo_Destino, ICAO_Aerodromo_Destino, ICAO_Empresa_Aerea
""").createOrReplaceTempView("tb_quantidade_destino")


#UNINDO AS TABELAS DE ORIGEM E DESTINO E CALCULANDO O TOTAL DE POUSOS E DECOLAGENS DE CADA AEROPORTO
spark.sql("""
    SELECT
        Nome_Aerodromo_Origem, ICAO_Aerodromo_Origem, tb_quantidade_partidas.ICAO_Empresa_Aerea AS ICAO_Empresa_Aerea_Origem, tb_quantidade_partidas.quantidade_partidas_do_aeroporto, 
        Nome_Aerodromo_Destino, ICAO_Aerodromo_Destino, tb_quantidade_destino.ICAO_Empresa_Aerea AS ICAO_Empresa_Aerea_Destino, tb_quantidade_destino.quantidade_destino_do_aeroporto, 
        (tb_quantidade_partidas.quantidade_partidas_do_aeroporto + tb_quantidade_destino.quantidade_destino_do_aeroporto) AS Total_Pousos_Decolagens,
        CONCAT(ICAO_Aerodromo_Origem, (tb_quantidade_partidas.quantidade_partidas_do_aeroporto + tb_quantidade_destino.quantidade_destino_do_aeroporto)) AS Chave
    FROM
        tb_quantidade_partidas
    LEFT JOIN
        tb_quantidade_destino ON tb_quantidade_partidas.ICAO_Aerodromo_Origem = tb_quantidade_destino.ICAO_Aerodromo_Destino
    WHERE
        tb_quantidade_partidas.ICAO_Empresa_Aerea = tb_quantidade_destino.ICAO_Empresa_Aerea
    ORDER BY
        ICAO_Aerodromo_Origem
""").createOrReplaceTempView("tb_qtd_pousos_decolagens_tmp")


#SELECIONANDO COMPANHIA AEREA COM MAIOR ATUACAO PARA CADA AEROPORTO
spark.sql("""
    SELECT
        Nome_Aerodromo_Origem AS Nome_Aerodromo, ICAO_Aerodromo_Origem AS ICAO_Aerodromo, MAX(Total_Pousos_Decolagens) AS Total_Pousos_Decolagens,
        CONCAT(ICAO_Aerodromo_Origem, MAX(Total_Pousos_Decolagens)) AS Chave
    FROM
        tb_qtd_pousos_decolagens_tmp
    GROUP BY
        Nome_Aerodromo_Origem, ICAO_Aerodromo_Origem
""").createOrReplaceTempView("tb_qtd_pousos_decolagens")



#JOIN ENTRE A TABELA QUE POSSUI A CHAVE COM O VALOR TOTAL DE POUSOS E DECOLAGENS MAXIMO DE CADA AEROPORTO COM A TABELA QUE POSSUI ESSE MESMO VALOR PARA CADA AEROPORTO POREM COM TODAS COMPANHIAS AEREAS
#FAZENDO O LEFT JOIN, PEGAMOS SOMENTE A COMPANHIA AEREA QUE POSSUI O VALOR MAXIMO DE POUSOS E DECOLAGEM EM CADA AEROPORTO 
spark.sql("""
    SELECT 
        Nome_Aerodromo, ICAO_Aerodromo, tb_qtd_pousos_decolagens_tmp.ICAO_Empresa_Aerea_Origem AS ICAO_Empresa_Aerea, tb_qtd_pousos_decolagens_tmp.quantidade_partidas_do_aeroporto,
        tb_qtd_pousos_decolagens_tmp.quantidade_destino_do_aeroporto, tb_qtd_pousos_decolagens.Total_Pousos_Decolagens
    FROM
        tb_qtd_pousos_decolagens
    LEFT JOIN
        tb_qtd_pousos_decolagens_tmp ON tb_qtd_pousos_decolagens.Chave = tb_qtd_pousos_decolagens_tmp.Chave
    ORDER BY
        ICAO_Aerodromo
""").createOrReplaceTempView("tb_qtd_pousos_decolagens_max")



#JOIN COM A TABELA DE AIRCIA PARA MOSTRARMOS A VIEW FILTRADA COM OS DADOS DE AEROPORTOS DISPONIVEIS NA AIRCIA

df_view2 = spark.sql("""
    SELECT
        Nome_Aerodromo, ICAO_Aerodromo, aircia.Razao_Social AS Razao_Social_CIA_Aerea, quantidade_partidas_do_aeroporto, quantidade_destino_do_aeroporto, Total_Pousos_Decolagens
    FROM
        tb_qtd_pousos_decolagens_max
    RIGHT JOIN
        aircia ON tb_qtd_pousos_decolagens_max.ICAO_Empresa_Aerea = aircia.ICAO
    ORDER BY
        ICAO_Aerodromo
""")


#VIEW 2

#EMPRESAS AEREAS COM DADOS DE VIAGENS NULOS DEVEM AO FATO DE NAO TEREM DADOS NA TABELA DE VRA
#ALGUNS AEROPORTOS PODEM VIR COM O NOME NULO PELO FATO DE NAO TEREM ESTE DADO FORNECIDO PELA API

df_view2.show()
```
