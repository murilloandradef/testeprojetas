```python
def ordernar_lista_decrescente(lista):
   
    lista_decrescente = sorted(lista, reverse=True)
    return lista_decrescente
```

**Desenvolva um código python para exclusão dos itens duplicados da lista acima.**

```python
def remover_duplicados(lista):
    
    lista_sem_duplicados = []
    
    for i in lista:
      if i not in lista_sem_duplicados:
        lista_sem_duplicados.append(i)
        
    return lista_sem_duplicados
```

```python
remover_duplicados(lista_repetida)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import (
    StructType,
    StructField, 
    StringType, 
    IntegerType, 
    DoubleType,
    DateType
)
```

```python
# Data and Structure
from datetime import date

clientes_schema = StructType([
    StructField("codigo", IntegerType(),True),
    StructField("nome", StringType(),True),
    StructField("telefone", StringType(),True),
    StructField("uf", StringType(), True),
    StructField("municipio", StringType(), True),
])

clientes_data = [
    (1, "James Smith", "+11 (11) 11111-1111", "SP", "São Paulo"),
    (2, "Michael Rose", "+22 (22) 22222-2222", "SP", "Campinas"),
    (3, "Robert Williams", "+33 (33) 33333-3333", "RJ", "Rio de Janeiro"),
    (4, "Maria Anne Jones", "+44 (44) 44444-4444", "MG", "Belo Horizonte"),
    (5, "Jen Mary Brown", "+55 (55) 55555-5555", "MG", "Divinópolis"),
]

produtos_schema = StructType([
    StructField("codigo", IntegerType(), True),
    StructField("descricao", StringType(), True),
    StructField("valor", DoubleType(), True),
])

produtos_data = [
    (1, "Caneta Simples", 12.5),
    (2, "Caderno Simples", 55.0),
    (3, "Folha avulsa", 1.0),
    (4, "Borracha", 2.0),
]

pedidos_produtos_schema = StructType([
    StructField("codigo", IntegerType(), True),
    StructField("codigo_cliente", IntegerType(), True),
    StructField("codigo_produto", IntegerType(), True),
    StructField("quantidade", IntegerType(), True),
])

pedidos_produtos_data = [
    (1, 1, 1, 50),
    (2, 2, 2, 150),
    (2, 2, 1, 75),
    (3, 2, 1, 200),
    (1, 1, 4, 10),
]

pedidos_schema = StructType([
    StructField("codigo", IntegerType(), True),
    StructField("data", DateType(), True),
    StructField("total", DoubleType(), True),
])

pedidos_data = [
    (1, date.fromisoformat("2022-01-31"), 8.0)
]
```

```python
# Get Spark Session
spark = SparkSession.builder.master("local[*]").getOrCreate()

df_clientes = spark.createDataFrame(data=clientes_data, schema=clientes_schema)
df_clientes.createOrReplaceTempView("clientes")

df_produtos = spark.createDataFrame(data=produtos_data, schema=produtos_schema)
df_produtos.createOrReplaceTempView("produtos")

df_pedidos_produtos = spark.createDataFrame(data=pedidos_produtos_data, schema=pedidos_produtos_schema)
df_pedidos_produtos.createOrReplaceTempView("pedidos_produtos")

df_pedidos = spark.createDataFrame(data=pedidos_data, schema=pedidos_schema)
df_pedidos.createOrReplaceTempView("pedidos")
```

**Exercício 4**

**Utilizando linguagem SQL e as estruturas de dados criadas acima, desenvolva um script para listar o valor total do pedido por cliente e produto.**

```python
df_valor_total = spark.sql("""
  SELECT
      nome_cliente, nome_produto, SUM((valor_unitario_produto*quantidade)) AS valor_total
  FROM
      (
        SELECT 
        clientes.nome AS nome_cliente, produtos.descricao AS nome_produto, produtos.valor AS valor_unitario_produto, pedidos_produtos.quantidade
      FROM
        pedidos_produtos
      LEFT JOIN
        clientes ON pedidos_produtos.codigo_cliente = clientes.codigo
      LEFT JOIN
        produtos ON pedidos_produtos.codigo_produto = produtos.codigo
      )
  GROUP BY
      nome_cliente, nome_produto
  ORDER BY
      nome_cliente
""")

df_valor_total.createOrReplaceTempView("tb_valor_total")
df_valor_total.show()
```

**Utilizando linguagem pyspark e as estruturas de dados criadas acima, desenvolva um código para classificar os clientes como VIP ou PADRAO.**

VIP - cliente com valor total de pedidos superior ou igual a R$1.000,00

PADRAO - cliente com valor total de pedidos inferior a R$1.000,00 ou que não tenha nenhum pedido solicitado

```python
from pyspark.sql.functions import *
```

```python
df_valor_total_final = df_valor_total.drop("nome_produto").groupBy("nome_cliente").agg(sum("valor_total").alias("valor_total"))


df_clientes.join(df_valor_total_final, df_clientes.nome == df_valor_total_final.nome_cliente, "left").withColumn("classificacao", when(df_valor_total_final.valor_total >= 1000, 'VIP').otherwise('PADRAO')).show() 
```
